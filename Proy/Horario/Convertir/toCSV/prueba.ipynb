```python
import os
import pandas as pd

# Directorio de entrada (donde están los archivos .xlsx)
input_directory = 'D:/Algo/Win/VSC/pocotu/tod/Proy/Horario/Convertir/toCSV/XLSXs'

# Directorio de salida (donde se guardarán los archivos .csv)
output_directory = 'D:/Algo/Win/VSC/pocotu/tod/Proy/Horario/Convertir/toCSV/CSVs'

# Obtener la lista de archivos .xlsx en el directorio de entrada
xlsx_files = [f for f in os.listdir(input_directory) if f.endswith('.xlsx')]

# Recorrer cada archivo .xlsx y convertirlo a .csv
for xlsx_file in xlsx_files:
    # Crear el nombre completo de archivo de entrada y salida
    input_file_path = os.path.join(input_directory, xlsx_file)
    output_file_path = os.path.join(output_directory, os.path.splitext(xlsx_file)[0] + '.csv')
    
    # Leer el archivo .xlsx y guardarlo como .csv en el directorio de salida
    df = pd.read_excel(input_file_path)
    df.to_csv(output_file_path, index=None, header=True)

    print(f"Convertido '{input_file_path}' a '{output_file_path}'")

print("Proceso completado.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos de ejemplo
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2, 3.4, 3.8, 5.5, 6.7, 8.2])

# Crear un modelo de regresión lineal
model = LinearRegression()

# Ajustar el modelo a los datos
model.fit(x.reshape(-1, 1), y)

# Predecir valores usando el modelo
y_pred = model.predict(x.reshape(-1, 1))

# Graficar los datos y la regresión
plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x, y_pred, color='red', label='Regresión lineal')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión Lineal Simple')
plt.legend()
plt.show()
```
