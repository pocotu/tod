```python
# Libreria para analisis de datos
import numpy as np
# Libreria para graficar
import matplotlib.pyplot as plt
# Libreria para importar datos
from sklearn.datasets import load_boston
```

```python
# cargando libreria de datos
boston = load_boston()
print(boston.DESCR)
```

##### Promedio de habitaciones por vivienda en Boston

```python
X = boston.data[:,5]
print(X)
```

##### Valor medio de las viviendas en Boston

```python
Y = boston.target
print(Y)
```

Formula Minimizar el error cuadratico medio (MCO): $$\beta = (X^TX)^{-1}X^TY$$ 

##### Convertimos los datos en arrays e imprimimos la grafica

```python
X = np.array(boston.data[:,5])
Y = np.array(boston.target)
plt.scatter(X,Y,alpha=0.3) # alpha es la transparencia de los puntos
```

```python
# Aadimos columna de 1s para termino independiente y transponemos la matriz
X = np.array([np.ones(506),X]).T
# "X.T" Imprime de forma transpuesta "X" para que no ocupe tanta pantalla
print(X.T)
```

##### Escritura de la funcion de minimizacion: $\beta = (X^TX)^{-1}X^TY$

```python
# linalg.inv(es parte del paquete de Algebra Lineal), calcula la inversa de una matriz
B = np.linalg.inv(X.T @ X) @ X.T @ Y
print(B)
```

```python
a, b = B
print(f"Donde {a} es el punto de corte en el eje Y cuando X=0")
print(f"y {b} es el incremento del precio de la vivienda por cada unidad de incremento en el numero medio de habitaciones")
print("El modelo de regresion lineal es: Y = {:.4f} + {:.4f}X".format(a,b))
```

#### Graficando la recta de regresion

```python
# Inserta la recta de regresion desde 4 hasta 9 habitaciones
plt.plot([4,9], [B[0]+B[1]*4, B[0]+B[1]*9], c="red")
# Inserta los puntos de la grafica
plt.scatter(X[:,1],Y,alpha=0.3)
```
