```python
import scipy.stats as stats

# Define the mean and standard deviation
mu = 200
sigma = 15

# Calculate the Z scores for 220ms and 185ms and 215ms
z1 = (220 - mu) / sigma
print(z1)
z2 = (185 - mu) / sigma
print(z2)
z3 = (215 - mu) / sigma
print(z3)

# Calculate the probabilities
p1 = 1 - stats.norm.cdf(z1)  # P(X > 220)
print("p1", p1)
p2 = stats.norm.cdf(z3) - stats.norm.cdf(z2)  # P(185 < X < 215)
print("p2", p2)

# Print the results
print(f"The probability that a randomly selected response time is greater than 220ms is {p1:.4f}")
print(f"The probability that a randomly selected response time is between 185ms and 215ms is {p2:.4f}")

```
