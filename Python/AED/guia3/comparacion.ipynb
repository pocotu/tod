```python
import random
import time
import matplotlib.pyplot as plt
```

**Metodo Quicksort**

```python
def quicksort(lista):
    if len(lista) <= 1:
        return lista
    pivot = lista[len(lista) // 2]
    izquierda = [x for x in lista if x < pivot]
    medio = [x for x in lista if x == pivot]
    derecha = [x for x in lista if x > pivot]
    return quicksort(izquierda) + medio + quicksort(derecha)
```

**Metodo Burbuja, el mas eficiente**

```python
def ordenamiento_burbuja_optimizado2(lista):
    n = len(lista)
    intercambio = True
    iteracion = 0

    while (intercambio):
        intercambio = False
        for i in range(n - iteracion - 1):
            if (lista[i] > lista[i+1]):
                temp1 = lista[i]
                lista[i] = lista[i+1]
                lista[i+1] = temp1
                #lista[j], lista[j+1] = lista[j+1], lista[j]
                intercambio = True
        iteracion += 1
    #print("Numero de iteracion: ", iteracion)
    return lista
```

**Metodo Insercion**

```python
def ordenamiento_insercion(A):
    n = len(A)

    for i in range(n):
        posicion = i
        aux = A[i]
        while (posicion > 0) and (A[posicion-1] > aux):
            A[posicion] = A[posicion-1]
            posicion = posicion - 1
        A[posicion] = aux
    return A
```

**Impresion de listas ordenadas**

```python
numero = 500
A = []
for i in range(numero):
    A.append(random.randint(0, numero))

print("Lista original: ", A)
print("Lista ordenada por Quicksort")
print(quicksort(A))
print("Lista ordenada por Insercion")
print(ordenamiento_insercion(A))
print("Lista ordenada por Burbuja")
print(ordenamiento_burbuja_optimizado2(A))
```

**Grafica de comparacion de los tiempos de ejecucion de los algoritmos de ordenamiento**

```python
# tiempo de ejecucion de los algoritmos
t_quicksort = []
t_insercion = []
t_burbuja = []
# lista de n elementos aleatorios de 
n = []

for i in range(A[0], A[len(A)-1]):
    # lista de n elementos aleatorios de 

    n.append(i)
    lista = random.sample(range(1, 1001), i)
    
    start_time = time.time()
    quicksort(lista)
    t_quicksort.append(time.time() - start_time)
    #print("Lista ordenada por Insercion")
    start_time = time.time()
    ordenamiento_insercion(lista)
    t_insercion.append(time.time() - start_time)
    #print("Lista ordenada por Burbuja")
    start_time = time.time()
    ordenamiento_burbuja_optimizado2(lista)
    t_burbuja.append(time.time() - start_time)

plt.plot(n, t_quicksort, label = "Quicksort")
plt.plot(n, t_insercion, label = "Insercion")
plt.plot(n, t_burbuja, label = "Burbuja")
plt.xlabel('n')
plt.ylabel('tiempo')
plt.title('Comparacion de tiempos')
plt.legend()
plt.show()
```
